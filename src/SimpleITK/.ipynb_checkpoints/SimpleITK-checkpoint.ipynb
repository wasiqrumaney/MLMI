```python
def sitk_show(img, title=None, margin=0.05, dpi=40 ):
    nda = sitk.GetArrayFromImage(img)
    spacing = img.GetSpacing()
    figsize = (1 + margin) * nda.shape[0] / dpi, (1 + margin) * nda.shape[1] / dpi
    extent = (0, nda.shape[1]*spacing[1], nda.shape[0]*spacing[0], 0)
    fig = plt.figure(figsize=figsize, dpi=dpi)
    ax = fig.add_axes([margin, margin, 1 - 2*margin, 1 - 2*margin])

    plt.set_cmap("gray")
    ax.imshow(nda,extent=extent,interpolation=None)
    
    if title:
        plt.title(title)
    
    plt.show()
```

# Resampling

```python
def resample(image, new_spacing=[1,1,1]):
    # Determine current pixel spacing
    #spacing = np.array([scan[0].SliceThickness] + scan[0].PixelSpacing, dtype=np.float32)
    spacing = np.array(image.GetSpacing(), dtype=np.float32)
    
    image_shape=np.array(image.GetSize(),dtype=np.float32)
    
    resize_factor = spacing / new_spacing
    new_real_shape = image_shape * resize_factor
    new_shape = np.round(new_real_shape)
    real_resize_factor = new_shape / image_shape
    new_spacing = spacing / real_resize_factor
    print(resize_factor,new_real_shape,new_shape,real_resize_factor,new_spacing)
    nda = sitk.GetArrayFromImage(image)
    image = scipy.ndimage.interpolation.zoom(nda, real_resize_factor, mode='nearest')
    
    return image, new_spacing
```

```python
resampled, spacing = resample(img, [1,1,1])
```
